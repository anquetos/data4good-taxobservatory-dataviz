# Data For Good - Tax Observatory dataset EDA

Perform a few exploratory analysis to understand what data we are dealing with.

```python
# Import dataset in a DataFrame
df = pd.read_csv('data_final_dataviz.csv', sep=',')
```

```python
# Display sample of data
df.sample(5)
```

```python
# Display dimensions
print(f'Dimensions of the datset are :\n- {df.shape[0]} rows ;\n- {df.shape[1]} columns.')
```

```python
# Display columns names
df.columns
```

```python
# Display data types
df.dtypes
```

```python
# Calculate percentage of missing values per column
(df.isna().sum() / len(df) * 100).apply(lambda x: '{:.1f} %'.format(x))
```

```python
# Generate descriptive statistics
df.describe().round(2)
```

## What does the `empty` column mean ?

```python
# Display a sample of True 'empty' columns.
df[df['empty'] == True].sample(5)
```

It seems to be rows where all financial information are missing.

```python
# List 'financial' columns
cols = ['profit_before_tax', 'tax_accrued', 'tax_paid', 'employees',
       'unrelated_revenues', 'related_revenues', 'stated_capital',
       'accumulated_earnings', 'tangible_assets', 'total_revenues']

# Calculate the number of rows where all financial columns are NaN
print(f'Number of rows with no financial information : {len(df.loc[df[cols].isna().all(axis='columns')])}.')
```

```python
# Calculate the number of rows where 'empty' is True
print(f'Number of rows where "empty" columns is True : {len(df.loc[df['empty'] == True])}.')
```
